### 6. Creating a Categorical Plot

```python
# Creating a Categorical plot 
sns.histplot(df['Happiness Score'], bins = 20, kde = True)
```

```python
# Creating a Happiness flag for Countries
df.loc[df['Happiness Score'] < 4, 'Happiness Status'] = 'Unhappy Country'
df.loc[(df['Happiness Score'] >= 4) & (df['Happiness Score'] < 6), 'Happiness Status'] = 'Somewhat Happy Country'
df.loc[df['Happiness Score'] >= 6, 'Happiness Status'] = 'Happy Country'
```

```python
# Checking creation of Exclusion flag 
df['Happiness Status'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Creating a categorical plot in seaborn using the Happiness Status created above

sns.set(style="ticks")
g= sns.catplot(x="Country", y="Happiness Score", hue="Happiness Status", data=df)
```

#### This categorical plot displays that a majority of the countries in this study locate within the "Somewhat Happy Country" status. We see a few of countries in the "Unhappy Country" status and the rest of the countries landing within the "Happy Country status.
